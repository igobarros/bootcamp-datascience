## Import the relevant libraries

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
df = pd.read_csv('Countries-exercise.csv')
df.sample(5)
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlabel('Longitude', fontsize=16)
plt.ylabel('Latitude', fontsize=16)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
X = df.iloc[:, 1:]
```

## Clustering

Assume there are only two clusters. 

```python
kmeans = KMeans(8)
```

```python
kmeans.fit(X)
```

### Clustering Resutls

```python
results = kmeans.fit_predict(X)
```

```python
data_results = df.copy()
data_results['Clusters'] = results
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
plt.scatter(data_results['Longitude'], data_results['Latitude'], c=data_results['Clusters'], cmap='rainbow')
plt.xlabel('Longitude', fontsize=16)
plt.ylabel('Latitude', fontsize=16)
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```
